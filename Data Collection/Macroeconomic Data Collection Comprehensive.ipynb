# Macroeconomic Data Collection Pipeline
---
This notebook collects and compiles quarterly macroeconomic time series for the UK and US, to be used as candidate predictor variables in the macroeconomic model for Probability of Default (PD) prediction.

**Data Sources Used:**
- **FRED** (Federal Reserve Bank of St. Louis): `fredapi` Python library
- **OECD**: Direct API calls via `requests`
- **Yahoo Finance**: `yfinance` Python library

**Temporal Frequency:** Quarterly (all monthly/daily series are resampled to quarter end)

**Date Coverage:** All available data for each series

## 3. Helper Functions

```python
def resample_to_quarter_end(series):
    """
    Resample a time series to quarterly frequency using the last available
    observation in each quarter, anchored to the last day of the quarter.
    FRED dates observations to the start of the period but the value
    represents the end of that period, so we align to quarter-end.
    Strips timezone info to ensure consistency across sources.
    """
    # Strip timezone info (sources like Yahoo Finance attach UTC labels inconsistently)
    series.index = pd.to_datetime(series.index).tz_localize(None)

    # Take the last observation in each quarter, anchored to quarter-end dates
    # (e.g Q1 - 2025-03-31, Q2 - 2025-06-30)
    return series.resample('Q').last()


    
def fetch_oecd(url, col_name):
    """
    Fetch a single time series from the OECD SDMX REST API.
    Returns a two-column DataFrame: ['date', col_name].
    Adds a 1-second delay between calls to respect OECD rate limits.
    """
    # Request response as SDMX-CSV with human-readable labels
    headers = {'Accept': 'application/vnd.sdmx.data+csv; charset=utf-8; labels=both'}
    time.sleep(1)  # Avoid hitting OECD rate limits

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        df = pd.read_csv(StringIO(response.text))

        # Identify the time column dynamically — it starts with 'TIME_PERIOD'
        # but may have a label suffix (e.g. 'TIME_PERIOD: Time period')
        time_col = [c for c in df.columns if c.startswith('TIME_PERIOD')][0]

        # Keep only the time and value columns; drop metadata columns
        df_clean = df[[time_col, 'OBS_VALUE']].copy()
        df_clean.columns = ['date', col_name]

        # Coerce to numeric so non-numeric missing value codes become NaN
        df_clean[col_name] = pd.to_numeric(df_clean[col_name], errors='coerce')
        return df_clean
    else:
        # Return None so the pipeline continues rather than crashing
        print(f"WARNING: Could not fetch '{col_name}' - HTTP {response.status_code}")
        return None
```

## 4. FRED Data

Monthly and daily FRED series are resampled to quarterly frequency by taking the last observation of each quarter, anchored to the final day of the quarter (e.g. 2025-03-31). Quarterly series (GDP, credit) are reported by FRED with quarter-start dates but represent full-quarter values; their index is shifted to quarter-end to reflect this.

Growth rates are derived as follows:
- YoY growth: `pct_change(4)` - compares each quarter to the same quarter one year prior
- QoQ growth: `pct_change(1)` - compares each quarter to the immediately preceding quarter

### 4a. UK - FRED Series

(NOTE: the values of the most recent quarter may be inaccurate and misleading and will be dealt with at the end)

```python
# --- UK Real GDP (quarterly, seasonally adjusted) ---
# Source: FRED NGDPRSAXDCGBQ
uk_real_gdp_raw = fred.get_series('NGDPRSAXDCGBQ')
# FRED dates quarterly series to quarter-start; shift to quarter-end to reflect
# that the value represents the full quarter
uk_real_gdp_raw.index = uk_real_gdp_raw.index.to_period('Q').to_timestamp('Q')
uk_real_gdp       = uk_real_gdp_raw.copy()
uk_gdp_yoy_growth = uk_real_gdp_raw.pct_change(4) * 100  # YoY % growth

# --- UK Long-Term Government Bond Yield (10-year, monthly to quarterly) ---
# Source: FRED IRLTLT01GBM156N
uk_bond_yield_10y = resample_to_quarter_end(fred.get_series('IRLTLT01GBM156N'))

# --- UK Real Broad Effective Exchange Rate (monthly to quarterly) ---
# Source: FRED RBGBBIS
uk_reer = resample_to_quarter_end(fred.get_series('RBGBBIS'))

# --- UK Brent Crude Oil Price (daily to quarterly) ---
# Source: FRED DCOILBRENTEU
uk_oil_price = resample_to_quarter_end(fred.get_series('DCOILBRENTEU'))

# --- UK Total Credit to Non-Financial Sector, Adjusted for Breaks (quarterly) ---
# Source: FRED QGBCAMUSDA
uk_credit_raw = fred.get_series('QGBCAMUSDA')
# Shift index to quarter-end, same as GDP
uk_credit_raw.index  = uk_credit_raw.index.to_period('Q').to_timestamp('Q')
uk_credit            = uk_credit_raw.copy()
uk_credit_qoq_growth = uk_credit_raw.pct_change(1) * 100  # QoQ % growth
uk_credit_yoy_growth = uk_credit_raw.pct_change(4) * 100  # YoY % growth

# --- UK Industrial Production Index (monthly to quarterly) ---
# Source: FRED IPIUKM
uk_industrial_production = resample_to_quarter_end(fred.get_series('IPIUKM'))

# --- Compile UK FRED DataFrame ---
fred_uk = pd.DataFrame({
    'uk_real_gdp':              uk_real_gdp,
    'uk_gdp_yoy_growth':        uk_gdp_yoy_growth,
    'uk_bond_yield_10y':        uk_bond_yield_10y,
    'uk_reer':                  uk_reer,
    'uk_oil_price':             uk_oil_price,
    'uk_credit':                uk_credit,
    'uk_credit_qoq_growth':     uk_credit_qoq_growth,
    'uk_credit_yoy_growth':     uk_credit_yoy_growth,
    'uk_industrial_production': uk_industrial_production,
})

fred_uk.index = pd.to_datetime(fred_uk.index).tz_localize(None)
print(f"UK FRED data: {fred_uk.index.min().date()} to {fred_uk.index.max().date()}, {len(fred_uk)} rows")
```

## 5. Yahoo Finance - Equity Indices

Daily prices are resampled to quarterly frequency by taking the closing price of the last trading day of each quarter, consistent with the quarter-end convention used throughout this dataset.

A quarter-on-quarter (QoQ) growth rate is also computed, since equity index levels are non-stationary and the growth rate can be a useful input for modelling.

(NOTE: the values of the most recent quarter may be inaccurate and misleading and will be dealt with at the end)

```python
# --- UK: FTSE 100 ---
# Download full history of daily prices, then take the closing price of the last trading day of each quarter
ftse_raw           = yf.Ticker("^FTSE").history(start="1984-01-01", interval="1d")
ftse_close         = ftse_raw['Close'].resample('Q').last().rename('uk_ftse_close')
ftse_close.index   = ftse_close.index.tz_localize(None)
uk_ftse_qoq_growth = ftse_close.pct_change(1) * 100  # QoQ % change
uk_ftse_qoq_growth.name = 'uk_ftse_qoq_growth'

# --- US: S&P 500 ---
# Same approach as FTSE
sp500_raw           = yf.Ticker("^GSPC").history(start="1950-01-01", interval="1d")
sp500_close         = sp500_raw['Close'].resample('Q').last().rename('us_sp500_close')
sp500_close.index   = sp500_close.index.tz_localize(None)
us_sp500_qoq_growth = sp500_close.pct_change(1) * 100  # QoQ % change
us_sp500_qoq_growth.name = 'us_sp500_qoq_growth'

print(f"FTSE 100: {ftse_close.index.min().date()} to {ftse_close.index.max().date()}, {len(ftse_close)} quarters")
print(f"S&P 500:  {sp500_close.index.min().date()} to {sp500_close.index.max().date()}, {len(sp500_close)} quarters")
```

```python
def build_oecd_dataframe(queries, monthly_cols=None):
    """
    Fetch all OECD series defined in `queries`, handle date parsing,
    and join into a single DataFrame indexed by quarter-end timestamps.

    `monthly_cols`: list of column names that are monthly (YYYY-MM format).
    These are resampled to quarter-end before joining.
    All other columns are assumed quarterly (YYYY-Qn format).
    """
    if monthly_cols is None:
        monthly_cols = []

    frames = []
    for col_name, url in queries.items():
        df = fetch_oecd(url, col_name)
        if df is None:
            continue

        if col_name in monthly_cols:
            # Monthly series: parse as datetime, resample to quarter-end using last observation
            df['date'] = pd.to_datetime(df['date'])
            df = df.set_index('date').resample('Q').last()
            df.index = df.index.normalize()  # strip sub-day precision
        else:
            # Quarterly series: convert YYYY-Qn period strings to quarter-end timestamps
            df['date'] = pd.PeriodIndex(df['date'], freq='Q').to_timestamp(how='end')
            df = df.set_index('date')
            df.index = df.index.normalize()  # strip sub-day precision

        frames.append(df)

    result = pd.concat(frames, axis=1)
    result.index = pd.to_datetime(result.index).tz_localize(None)
    return result


# Fetch UK OECD data
oecd_uk = build_oecd_dataframe(
    oecd_queries_uk,
    monthly_cols=['uk_consumer_confidence']
)

# Fetch US OECD data
oecd_us = build_oecd_dataframe(
    oecd_queries_us,
    monthly_cols=['us_consumer_confidence']
)

print(f"OECD UK: {oecd_uk.index.min().date()} to {oecd_uk.index.max().date()}, {len(oecd_uk)} rows")
print(f"OECD US: {oecd_us.index.min().date()} to {oecd_us.index.max().date()}, {len(oecd_us)} rows")
```

## 7. Combine into Final Datasets

All sources are joined on the datetime index using an outer join to preserve the full history of each series. Some rows will have NaNs where a given series starts later or ends earlier than others - this is expected and will be addressed during preprocessing.

All indices are normalised to midnight timestamps before joining to ensure exact date alignment. The current incomplete quarter is dropped at the end, as its values do not yet represent a full quarter.

```python
# Normalise all indices to midnight to ensure exact date alignment across sources
for df in [fred_uk, fred_us, oecd_uk, oecd_us]:
    df.index = pd.to_datetime(df.index).normalize()

ftse_close.index          = pd.to_datetime(ftse_close.index).normalize()
uk_ftse_qoq_growth.index  = pd.to_datetime(uk_ftse_qoq_growth.index).normalize()
sp500_close.index         = pd.to_datetime(sp500_close.index).normalize()
us_sp500_qoq_growth.index = pd.to_datetime(us_sp500_qoq_growth.index).normalize()

# --- UK Master Dataset ---
macro_uk = oecd_uk.join(
    [fred_uk, ftse_close, uk_ftse_qoq_growth],
    how='outer'
).sort_index()

# --- US Master Dataset ---
macro_us = oecd_us.join(
    [fred_us, sp500_close, us_sp500_qoq_growth],
    how='outer'
).sort_index()

# Drop the current incomplete quarter — values do not yet represent a full quarter
current_quarter_end = pd.Timestamp.today().to_period('Q').to_timestamp('Q').normalize()
macro_uk = macro_uk[macro_uk.index < current_quarter_end]
macro_us = macro_us[macro_us.index < current_quarter_end]

print("UK dataset shape:", macro_uk.shape)
print("US dataset shape:", macro_us.shape)
```

## 8. Coverage Audit

Check the start date, end date, and number of non-null observations for each variable. This gives a picture of which series have less historical data and what the effective modelling window will be.

```python
def coverage_summary(df, label):
    """
    Print a table showing, for each column: first non-null date,
    last non-null date, and count of non-null observations.
    """
    print(f"\n{'='*60}")
    print(f"  Coverage Summary: {label}")
    print(f"{'='*60}")
    summary = pd.DataFrame({
        'First Obs':  df.apply(lambda c: c.first_valid_index()),
        'Last Obs':   df.apply(lambda c: c.last_valid_index()),
        'N Non-Null': df.notna().sum(),
        'N Missing':  df.isna().sum(),
    })
    print(summary.to_string())
    
    # Number of rows with no missing values across all columns
    complete_rows = df.dropna().shape[0]
    print(f"\n  Complete rows (no missing values): {complete_rows} of {len(df)}")

coverage_summary(macro_uk, "UK Macroeconomic Variables")
coverage_summary(macro_us, "US Macroeconomic Variables")
```
